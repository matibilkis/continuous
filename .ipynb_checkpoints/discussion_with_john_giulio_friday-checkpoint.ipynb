```python
def euler_step(x,t,dt,f,g,parameters):
      w = np.random.normal()
      return dt*f(t,x,parameters) + np.sqrt(dt)*w*g(t,x,parameters)
```

```python
def ff(x, t,parameters=None):
    return np.zeros(1)

def gg(x, t,parameters=None):
    return np.eye(1)


def rk4_steep( x, t, dt, f, g, parameters,w):
    a21 =   0.66667754298442
    a31 =   0.63493935027993
    a32 =   0.00342761715422
    a41 = - 2.32428921184321
    a42 =   2.69723745129487
    a43 =   0.29093673271592
    a51 =   0.25001351164789
    a52 =   0.67428574806272
    a53 = - 0.00831795169360
    a54 =   0.08401868181222

    q1 = 3.99956364361748
    q2 = 1.64524970733585
    q3 = 1.59330355118722
    q4 = 0.26330006501868

    t1 = t
    x1 = x
    k1 = dt*f( t1, x1, parameters) + np.sqrt(q1)*np.dot(w,g( t1, x1, parameters))

    t2 = t1 + (a21*dt)
    x2 = x1 + (a21*k1)
    k2 = dt*f(t2, x2, parameters) + np.sqrt(q2)*np.dot(w,g(t2, x2, parameters))

    t3 = t1 + (a31 + a32)*dt
    x3 = x1 + (a31*k1) + (a32*k2)
    k3 = dt*f(t3,x3,parameters) + np.sqrt(q3)*np.dot(w,g(t3,x3,parameters))

    t4 = t1 + (a41 * dt)  + (a42 * dt)  + (a43 * dt)
    x4 = x1 + (a41 * k1) + (a42 * k2) + (a43 * k3)

    k4 = dt*f(t4,x4,parameters) + np.sqrt(q4)*np.dot(w,g(t4,x4,parameters))

    xstar = x1 + (a51*k1) + (a52*k2) + (a53*k3) + (a54*k4)

    return xstar
```

```python
ppo = [np.array([0])]
parameters=None
times = np.linspace(0,1,1000)
np.random.seed(seed)
dt=1/len(times)
for i in tqdm(range(len(times))):
    w=np.random.random(1)*np.sqrt(dt)
    ppo.append(rk4_steep(ppo[-1], times[i], dt, ff, gg, parameters, w))
```
