```python

def load(itraj = 1,total_time = 50., dt = 1e-3, exp_path=""):
    pp = get_path_config(total_time=total_time, dt=dt, itraj=itraj, exp_path = exp_path)
    states = np.load(pp+"states.npy")
    dys = np.load(pp+"dys.npy")
    return states, dys
```

```python
params, exp_path = def_params()
```

```python
ss=[]
dyy=[]
for k in range(1,6):
    states, dys = load(itraj=k, exp_path=exp_path, total_time=4., dt=1e-3)
    ss.append(states[:,0])
    dyy.append(dys[:,0])
```

```python
plt.figure(figsize=(7,3))
ax0,ax1 = plt.subplot(121),plt.subplot(122)
for k in range(len(ss)):
    ax0.plot(ss[k])
    ax1.plot(dyy[k])
```

```python
xi, kappa, omega, eta = params
```

```python
def give_matrices(xi, kappa, omega, eta):
    A = np.array([[-(xi + .5*kappa), omega], [-omega, xi - 0.5*kappa]])
    D = kappa*np.eye(2)
    E = B = -np.sqrt(eta*kappa)*np.array([[1.,0.],[0.,0.]])
    return A, D, E, B

```
